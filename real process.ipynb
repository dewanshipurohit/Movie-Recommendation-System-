```python
movies['tags']=movies['title']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
movies[movies['title']=="Jawan"].index[0]
```

```python
list(enumerate(similarity[0]))[0:5]
```

```python
movies['tags']=movies['title']
```

```python
movies['tags']=movies['title']+" "+movies['year'].astype(str)
```

```python
vectors=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
similarity=cosine_similarity(vectors)
```

```python
list(enumerate(similarity[0]))[0:10]
```

```python
distances=similarity[0]
```

```python
movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
```

```python
for i in movies_list:
    print(movies.iloc[i[0]].title)
```
